```python
import pandas as pd
import numpy as np
from datetime import timedelta
import os
```

```python
inputpath = "C:/Users/ehniv/OneDrive/SFU/2021 - 2022 Academics/Spring 2022/MATH 402W - Operations Research Clinic/datasets/Testing Datasets/"
filename = "20220304_location_4410_"

timeslot = pd.read_csv(os.path.join(inputpath, filename) + "time_slot.csv", index_col = 0)
timematrix = pd.read_csv(os.path.join(inputpath, filename) + "time_matrix.csv", index_col = 0)
```

# Processes time slot data

```python
morning = timeslot[timeslot["start_time"] < '14:00:00']
afternoon = timeslot[timeslot["start_time"] > '14:00:00']

morning.shape[0] + afternoon.shape[0] == timeslot.shape[0]
```

```python
morning_start = timedelta(hours = 6, minutes = 45)
morning_end = timedelta(hours = 14, minutes = 0)

depot = pd.DataFrame({'dummy_add':'Add_0', 'start_time':morning_start, 'end_time':morning_end,'start_time_scaled':0, 'end_time_scaled':0 }
                    , index = [0])


morning = pd.concat([depot, morning], ignore_index = True)
morning["start_time_scaled"] = (morning["start_time"] - morning_start).apply(lambda x: x.total_seconds()/60)
morning["end_time_scaled"] = (morning["end_time"] - morning_start).apply(lambda x: x.total_seconds()/60)
morning.head(n = 5)
```

```python
afternoon_start = timedelta(hours = 14, minutes = 45)
afternoon_end = timedelta(hours = 22, minutes = 0)

depot = pd.DataFrame({'dummy_add':'Add_0', 'start_time':afternoon_start, 'end_time':afternoon_end,'start_time_scaled':0, 'end_time_scaled':0 }
                    , index = [0])


afternoon = pd.concat([depot, afternoon], ignore_index = True)
afternoon["start_time_scaled"] = (afternoon["start_time"] - afternoon_start).apply(lambda x: x.total_seconds()/60)
afternoon["end_time_scaled"] = (afternoon["end_time"] - afternoon_start).apply(lambda x: x.total_seconds()/60)
afternoon.head(n = 5)
```

# Saves the Data

```python
morning.to_csv(os.path.join(inputpath,'Processed',filename + 'morning_time_slot.csv'))
afternoon.to_csv(os.path.join(inputpath,'Processed',filename + 'afternoon_time_slot.csv'))
```

# Process Time Matrix

```python
morning_locations = list(morning['dummy_add'])

morning_timematrix = timematrix[timematrix.index.isin(morning_locations)][morning_locations]
morning_timematrix
```

```python
afternoon_locations = list(afternoon['dummy_add'])

afternoon_timematrix = timematrix[timematrix.index.isin(afternoon_locations)][afternoon_locations]
afternoon_timematrix
```

# Saves the Time Matrix

```python
morning_timematrix.to_csv(os.path.join(inputpath,'Processed',filename + 'morning_time_matrix.csv'))
afternoon_timematrix.to_csv(os.path.join(inputpath,'Processed',filename + 'afternoon_time_matrix.csv'))
```
